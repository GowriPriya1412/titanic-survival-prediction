# 23. Survival Rate by Fare Brackets

```python
fare_bins = pd.qcut(df['Fare'], 4)
survival_by_fare = df.groupby(fare_bins)['Survived'].mean()
survival_by_fare.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Survival Rate by Fare Quartiles')
plt.xlabel('Fare Range')
plt.ylabel('Survival Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 25.Family size influence on Survival

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
sns.barplot(data=df, x='FamilySize', y='Survived', palette='viridis')
plt.title('Survival Rate by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# 26. Title Extraction from names and its Effect on Survival
```

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
title_survival = df.groupby('Title')['Survived'].mean().sort_values(ascending=False)
title_survival.plot(kind='bar', color='purple')
plt.title('Survival Rate by Passenger Title')
plt.ylabel('Survival Rate')
plt.xticks(rotation=45)
plt.show()
```
